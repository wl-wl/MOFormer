```python
from rdkit import Chem
from rdkit.Chem import Draw

# 定义 Arginine (Arg) 的 SMILES 表示
arg_smiles = "C(CC(=O)O)CN=C(N)N"  # Arginine 的 SMILES

# 创建分子对象
arg_molecule = Chem.MolFromSmiles(arg_smiles)

# 绘制分子图
img = Draw.MolToImage(arg_molecule, size=(500, 500))  # 500x500 像素的图像

# 显示分子图
img.show()

# 保存为高分辨率图片 (300 DPI)
img.save("Arg_molecule.png", dpi=(300, 300))

print("Arg 分子图已保存为图片：Arg_molecule.png")
```

```python
import matplotlib.pyplot as plt
import re

# 示例txt内容
file_path = 'result.txt'

# 读取文件内容
with open(file_path, 'r') as file:
    data = file.read()


# 解析数据
y_labels = []
r2_values = []

for line in data.strip().split("\n"):
    # 提取括号内的内容
    match_y = re.search(r'\((.*?)\)', line)
    # 提取 r2 值
    match_r2 = re.search(r'r2=([\d.]+)', line)
    if match_y and match_r2:
        y_labels.append(match_y.group(0))  # 包括括号内容
        r2_values.append(float(match_r2.group(1)))
```

```python

plt.figure(figsize=(5, 14))
y_positions = range(len(y_labels[:50]))
import matplotlib.colors as mcolors
colors = ['#504794', '#3b7bb1', '#aad69e', '#f4a85f', '#c24146', '#891b39']  # 6种颜色
bounds = [0, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]  # 区间边界
cmap = mcolors.ListedColormap(colors)
norm = mcolors.BoundaryNorm(bounds, cmap.N)

# 绘制水平点图
for y, r2 in zip(y_positions, r2_values[:50]):
    plt.plot([0, r2], [y, y], color="gray", linewidth=0.5)  # 水平线
    plt.scatter(r2, y, color=cmap(norm(r2)), s=50)  # 点

# 设置刻度
plt.yticks(y_positions, y_labels[:50])
plt.xlabel("R2 Value",fontsize=12)
plt.ylabel("Neuron Coordinate",fontsize=12)
# plt.minorticks_on()


plt.tick_params(direction='out', length=6, width=2, labelsize=12)  # 设置刻度线方向、长度和宽度
plt.tick_params(which='both')  # 显示主刻度和次刻度线
plt.tick_params(axis='both', which='both', bottom=True, left=True)
# 移除右边框和上边框
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# 设置左边框和下边框的粗细
ax.spines['left'].set_linewidth(2)
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_linewidth(2)
ax.spines['bottom'].set_color('black')


# 移除网格
plt.grid(False)
plt.xlim(-0.1,2)
plt.ylim(-0.5, len(y_labels[:50]) - 0.5)

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import re
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 文本文件路径
file_path = 'result.txt'

# 读取文件内容
with open(file_path, 'r') as file:
    text = file.read()

# 提取 (y, z) 坐标和 R2 值
pattern = r"fixing \(\d+\,(\d+)\,(\d+)\) with .*, r2=([\d\.]+)"
matches = re.findall(pattern, text)
```

```python
matches_aac=matches[:20*5]
matches_maccs=matches[20*5:5*187]
matches_acp=matches[187*5:199*5]
```

```python

# 将匹配结果转换为包含整数坐标和浮点数 R2 值的列表
data = [(int(y), int(z), float(r2)) for y, z, r2 in matches_aac]
data
```

```python
data = [(int(y), int(z), float(r2)) for y, z, r2 in matches_aac]
# 获取 y 和 z 的最大值以定义热力图网格的大小
max_y = max(d[0] for d in data)
max_z = max(d[1] for d in data)

# 创建一个空的 R2 值网格
r2_grid = np.full((max_y + 1, max_z + 1), np.nan)

# 用 R2 值填充网格
for y, z, r2 in data:
    r2_grid[y, z] = r2
    
from matplotlib.colors import LinearSegmentedColormap

# 自定义调色板，从白色到浅橙色到深橙色bcc6dd

#"#9aa591", "#f7f7f7", "#5e7050"
#"#c7afa2", "#f7f7f7", "#966a50"
cmap = LinearSegmentedColormap.from_list("custom_palette", ["#c7afa2", "#f7f7f7", "#966a50"])

# 绘制热力图 0.005
plt.figure(figsize=(6, 12))
ax=sns.heatmap(r2_grid, fmt=".3f", cmap=cmap, vmin=0, vmax=1, cbar_kws={'label': '$R^2$ Value'},linewidths=0.005, linecolor='white')

colorbar = ax.collections[0].colorbar
colorbar.ax.tick_params(width=2)  # 设置刻度线的宽度
# colorbar.ax.set_linewidth(1.5)    # 设置 color bar 本身的边框宽度
colorbar.ax.set_aspect(30)       # 设置 color bar 的宽度，通过改变 aspect 来调整


plt.xlabel('Hidden Neurons',fontsize=14)
plt.ylabel('X_acc',fontsize=14)
plt.savefig('formule_aac_cla.png',dpi=300)
# plt.title('(0,x,')
plt.show()
```
