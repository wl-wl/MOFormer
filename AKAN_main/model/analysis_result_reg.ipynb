```python
import re
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 文本文件路径
file_path = 'result_reg.txt'

# 读取文件内容
with open(file_path, 'r') as file:
    text = file.read()

# 提取 (y, z) 坐标和 R2 值
pattern = r"fixing \(\d+\,(\d+)\,(\d+)\) with .*, r2=([\d\.]+)"
matches = re.findall(pattern, text)
```

```python
matches_aac=matches[:20*7]
matches_maccs=matches[20*7:7*187]
matches_acp=matches[187*7:199*7]
```

```python
matches_1=matches[:199*7]
matches_1[-1]
```

```python
matches_2=matches[199*7:]
matches_2
```

```python

# 将匹配结果转换为包含整数坐标和浮点数 R2 值的列表
data = [(int(y), int(z), float(r2)) for y, z, r2 in matches_aac]
data
```

```python
data = [(int(y), int(z), float(r2)) for y, z, r2 in matches_aac]
# 获取 y 和 z 的最大值以定义热力图网格的大小
max_y = max(d[0] for d in data)
max_z = max(d[1] for d in data)

# 创建一个空的 R2 值网格
r2_grid = np.full((max_y + 1, max_z + 1), np.nan)

# 用 R2 值填充网格
for y, z, r2 in data:
    r2_grid[y, z] = r2
    
from matplotlib.colors import LinearSegmentedColormap

# 自定义调色板，从白色到浅橙色到深橙色bcc6dd
cmap = LinearSegmentedColormap.from_list("custom_palette", ["#EFD6D1", "#C89C91"])

# 绘制热力图
plt.figure(figsize=(6, 12))
sns.heatmap(r2_grid, fmt=".3f", cmap=cmap, vmin=0, vmax=1, cbar_kws={'label': 'R2 Value'},linewidths=0.01, linecolor='white')
plt.xlabel('Hidden Neurons')
plt.ylabel('X_acc')
# plt.title('(0,x,')
plt.show()
```
